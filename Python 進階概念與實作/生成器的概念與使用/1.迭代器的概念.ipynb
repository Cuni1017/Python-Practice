# 2. 定義一個function，參數a和x，回傳 a, a * 2, a * 3, ... a * x

```python
def get_list(a, x):
    ret = []
    for i in range(1, x + 1):
        ret.append(a * i)
    return ret
```

```python
print(get_list(5, 10))
```

```python
# 列表解析式
def get_list(a, x):
    return [a * i for i in range(1, x + 1)]
```

```python
print(get_list(5, 10))
```

# 3. 當 a = 5, x = 10 時 求和？ 求積？

```python
sum(get_list(5, 10))
```

```python
prod = 1
for i in get_list(5, 10):
    prod *= i

print(prod)
```

# 5.來寫一個自己的迭代器吧！

```python
class MyIterator:
    def __init__(self, a, x):
        self.a = a
        self.x = x
        self.index = 1
    
    def __next__(self):
        if self.index > self.x:
            return -1
        ret = self.a * self.index
        self.index += 1
        return ret
        
```

```python
it = MyIterator(5, 10000)
```

```python
prod = 1
while True:
    value = it.__next__()
    if value == -1:
        break
    prod *= value

print(prod)
```

```python
it = MyIterator(5, 10000)

s = 0
while True:
    value = it.__next__()
    if value == -1:
        break
    s += value

print(s)
```
